```python
def Dutch_national_flag(arr):
    low,mid,high = 0,0,len(arr)-1
    while mid<= high:
        if arr[mid] == 0:
            arr[low],arr[mid]=arr[mid],arr[low]
            mid+=1
            low+=1
        elif arr[mid] == 1:
            mid+=1
        else:
            arr[high] ,arr[mid] =arr[mid],arr[high]
            high -=1
            return
arr = [4,5,6,7,8,14,64,57]
Dutch_national_flag(arr)    
```

```python
#Wap to print the sum of first n natural numbers

def sumn(n):
    if n==1:
        return 1
    smallOutput = sumn(n-1)
    return smallOutput +n

sumn(10)
```

```python
#WAP to find the sum of all elements of the array or list recursively

def sumarrl(arr):
    if len(arr) == 1:
        return arr[0]
    smallOutput = sumarrl(arr[1:])
    return smallOutput + arr[0]

sumarrl([4,5,6,7,8,9,14,1,54])
```
