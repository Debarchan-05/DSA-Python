```python
#Right angle triangle with spaces
n = int(input("enter a number"))

for i in range(1,n+1):
    for j in range(n-i):
        print(" " , end=" ")
    for k in range(i):
            print("*",end=" ")
    print()
            
```

```python
#increase in alphabets
n = int(input("Enter a alphabet"))
for i in range(1,n+1):
    for j in range(i):
        print(chr(j+65),end =" ")
    print()
```

```python
#Merging of the patterns
n = int(input("enter a number"))

for i in range(1,n+1):
    for j in range((n+1)-i):
        print("*" , end=" ")
    print()
for i in range(1,n+1):
    for j in range(i):
        print("*",end=" ")
    print()
```

```python
#linear search in python
lst = list(map(int,input("enter the number").split()))
key = int(input("Element that needs to be found"))
found = True
for i in range(len(lst)):
    if lst[i] == key:
        print("element found at index ",i)
        found = True
        break
else:
    found = False
    print("element not found")
```

```python
#pyramid of numbers
n = int(input("Enter an number"))
i=1
while i<=n:
    j=1
    while j<=(n-i):
        print(".",end=" ")
        j=j+1
    k=i
    while k<(2*i):
        print(k ,end =" ")
        k+=1
    i=i+1
    print()

```
