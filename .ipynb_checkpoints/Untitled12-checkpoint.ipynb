```python
# Palindrome sequence
str1 = input("Enter the string")
str2 = input("Enter the string")

def is_palindrome(s):
    return s ==s[::-1]
print(f"{str1} is palindrome  {is_palindrome(str1)}")
print(f"{str2} is palindrome  {is_palindrome(str2)}")


```

```python
num = int(input("enter a number"))
def fact_num(n):
    if n < 0:
        return "factorial not def for negative number"
    elif n == 1:
        return 1
    result = 1
    for i in range(1,n+1):
        result *= i
    return result

print(f"the factorial of {num} is {fact_num(num)}")
```

```python
#wap to count the occurence of a string

# dct = {}
# traverse the string
# check if the character is present inside the dct
# if present increase the count otherwise create the entry of the element starting from 1
```

```python
dct = {}
def count_char(n):
    for char in n:
        if char in dct:
            dct[char] +=1
        else:
            dct[char] =  1
            
count_char("hello")
print(dct)


    
        
```
