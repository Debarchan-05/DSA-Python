```python
#Binary Search
#Sorted List
lst = list(map(int,input("enter the list numbers").split()))
print(lst)
k= int(input("enter the key value to be searched"))
l = len(lst)
i=0
j = (l-1)
while i <=j:
    mid = i+(j-i)
    if lst[mid] == k:
        print(f"{k} found at Index found at {mid}")
        break
    elif lst[mid] < k:
        i = mid +1
    else:
        j = mid-1
```

```python
#Bubbble Sort
def bubbleSort(arr):
    for i in range(len(arr)):
        for j in range(len(arr)-i-1):
            if arr[j] > arr[j+1]:
                arr[j],arr[j+1]=arr[j+1],arr[j]
arr = [4,5,7,10,18,20,78,99,54,67]
bubbleSort(arr)

```
