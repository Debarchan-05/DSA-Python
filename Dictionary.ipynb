```python
#Count number of alphabets,digits,special character and vowels
s = input("Enter a string: ")

vowels = 0
consonants = 0
digits = 0
special_chars = 0

for ch in s:
    if ch.isalpha():
        if ch.lower() in 'aeiou':
            vowels += 1
        else:
            consonants += 1
    elif ch.isdigit():
        digits += 1
    else:
        special_chars += 1

print("Vowels:", vowels)
print("Consonants:", consonants)
print("Digits:", digits)
print("Special Characters:", special_chars)
```

```python
# Count the number of times the character occured in a string

def count_str(s,ch):
    count = 0
    for i in s:
        if i == ch:
            count+=1
    return count

stri = input("Enter the string")
chara= input("enter the charater")
print("Occurences:" , count_str(stri,chara))
    
```

```python
# Replace all the vowels by * 
def replace_vowel(s):
    repl_vow =""
    for ch in s:
        if ch.isalpha():
            if ch.lower() in 'aeiou':
                repl_vow += "*"
            else:
                repl_vow += ch
    print(repl_vow)
replace_vowel("Hello")

                
```

```python
# Function to check whether a string is palindrome or not
def is_palindrome(s):
    for i in s:
        if s == s[::-1]:
            print("Palindrome")
            return True
        else:
            print("Not Palindrome")
            return False
is_palindrome("mimansha")
            
```

```python
# Write a user defined function to convert a string with more than one word into title case string where string is passed as parameter.

def tittle_case(s):
    ch = s.split()
    result = ""
    for i in ch:
        result+= i[0].upper() + i[1:].lower()+ " "
    return result
tittle_case("hello mimansha")
```

```python
# Write a function deleteChar() which takes two parameters one is a string and other is a character. The function should create a new string after deleting all occurrences of the character from the string and return the new string
def deleteChar(s,ch):
    res=""
    for i in s:
        if i!= ch:
            res+=i
    return res

deleteChar("Hello","e")
```
