```python
# second largest element
arr= [4,7,8,2,3,6,10]
largest_arr = -1
second_largest = -1
for i in range(0 , len(arr)):
    if arr[i] > largest_arr:
        temp = largest_arr
        largest_arr = arr[i]
        if temp > second_largest:
            second_largest = temp
print(second_largest)

```

```python
# index equillibrium
arr = [1,2,2,9,3,2]
left_sum = 0
sum1 = sum(arr)
for i in range(len(arr)):
    right_sum = sum1 - left_sum - arr[i]
    if left_sum == right_sum:
        print(i)
        break
    left_sum += arr[i]
```

```python
# Person situation based on temperature
temp = float(input("Enter body temperature: "))

if temp >= 85.0 and temp <= 91.0:
    print("Serious Hypothermia")

if temp > 91.0 and temp < 95.0:
    print("Mild Hypothermia")

if temp >= 95.0 and temp <= 98.0:
    print("Normal Temperature")

if temp > 98.0 and temp <= 100.0:
    print("Mild Fever")

if temp > 100.0 and temp <= 105.0:
    print("High Fever")
```

```python
# Positive Even and Odd
n = int(input("Enter a number: "))

if n > 0 and n % 2 != 0:
    print("Number is Positive and Odd")

if n > 0 and n % 2 == 0:
    print("Number is Positive and Even")

if n < 0 and n % 2 != 0:
    print("Number is Negative and Odd")

if n < 0 and n % 2 == 0:
    print("Number is Negative and Even")
```

```python
# natural numbers using  for loop
n = int(input("Enter N: "))

for i in range(1, n+1):
    print(i, end=" ")
```

```python
# natural numbers using while loop
n = int(input("Enter N: "))

i = 1
while i <= n:
    print(i, end=" ")
    i += 1
```

```python
# print digits from right to left
n = int(input("Enter number: "))

while n > 0:
    digit = n % 10
    print(digit)
    n = n // 10
```

```python
# Factors of a number
n = int(input("Enter a number: "))

for i in range(1, n + 1):
    if n % i == 0:
        print(i)
```

```python
# Count the factors of a number
n = int(input("Enter a number: "))
count = 0

for i in range(1, n + 1):
    if n % i == 0:
        count += 1

print("Number of factors:", count)
```

```python
# Whether a number is prime or not
n = int(input("Enter a number: "))
count = 0

for i in range(1, n+ 1):
    if n % i == 0:
        count += 1

if count == 2:
    print("Prime Number")
else:
    print("Not a Prime Number")
```
